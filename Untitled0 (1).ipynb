```python
import pandas as pd
df = pd.read_excel('Employee Sample Data - A.xlsx')
print(df.head())
```

```python
# Step 2: Clean the data
print(df.shape)
print(df.isnull().sum())
df = df.dropna(subset=['Annual Salary', 'Age' , 'Department'])
print(df.shape)
```

```python
# Step 3: Modify the first 5 rows
df.loc[0:4, 'Age'] = [30, 35, 40, 45, 50]
df.loc[0:4, 'Annual Salary'] = [50000, 60000, 70000, 80000, 90000]
print(df.head())
```

```python
# Step 4: Row with largest salary
print(df[df['Annual Salary'] == df['Annual Salary'].max()])
```

```python
# Step 5: Group by department
print(df.groupby('Department')[['Age', 'Annual Salary']].mean())
```

```python
# Step 6: Group by department + ethnicity
print(df.groupby(['Department', 'Ethnicity'])['Age'].agg(['max', 'min']).join(df.groupby(['Department', 'Ethnicity'])['Annual Salary'].median().rename('Median Salary')))
```

```python
# Step 7: Save to new Excel file
df.to_excel('Employee_Clean.xlsx', index=False)
print('File saved!')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Correlation Heatmap
plt.figure(figsize=(8,5))
sns.heatmap(df[['Age','Annual Salary','Bonus %']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 2. Gender Distribution
plt.figure(figsize=(6,6))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
# 2. Gender Distribution
plt.figure(figsize=(6,6))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
# 3. Salary Distribution by Department
plt.figure(figsize=(10,6))
df.boxplot(column='Annual Salary', by='Department')
plt.title('Salary Distribution by Department')
plt.suptitle('')
plt.xticks(rotation=45)
plt.show()
```

**bold text**

```python
# 4. Age Distribution
plt.figure(figsize=(10,6))
df['Age'].plot(kind='hist', bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# 5. Average Salary by Department
plt.figure(figsize=(10,6))
df.groupby('Department')['Annual Salary'].mean().plot(kind='bar')
plt.title('Average Salary in each Department')
plt.ylabel('Annual Salary')
plt.xticks(rotation=45)
plt.show()

# 6. Number of Employees by Department
plt.figure(figsize=(10,6))
df['Department'].value_counts().plot(kind='bar')
plt.title('Number of Employees in each Department')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
